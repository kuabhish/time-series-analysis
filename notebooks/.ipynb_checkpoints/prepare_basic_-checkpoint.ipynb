## create a basic data

```python
x = [ 1,2,3,4,5,6,   1,2,3,4,5,6,  1,2,3,4,5,6,   1,2,3,4,5,6 ]
len(x)
```

```python
def generate_random(lb = -0.2, ub = 0.2):
    return (ub-lb)*(np.random.random()) + lb
```

```python
def datagen():
  x = [ 1,2,3,4,5,6 ]
  if len(x) >0:
    for y in x:
      yield y
      x.append(y)
```

```python
def single_exponential_smoothing(x , a=0.5):
  res = [ np.nan , x[0]]
  for i in range( 2 , len(x)+1 ):
    res.append( a*x[i-1] + (1-a)*res[i-1] )
  return res
```

```python
single_exponential_smoothing(np.arange(1,3))
```

```python

%matplotlib notebook

fig, ax = plt.subplots(figsize=(10,5))

original = deque(maxlen=100)
created = deque(maxlen=100)
good_points = {}
anomalous_points = {}

index = 0
for i in datagen():
    original.append(i+generate_random())
    created.append(single_exponential_smoothing(original)[-1])
    
    ax.clear()
#     plt.plot(original , 'y', label='original')
#     plt.plot(created , 'g', label = 'predicted')
    
    if np.abs(created[-1]- original[-1])>1:
        anomalous_points[str(index)] = original[-1]
    else:
        good_points[str(index)] = original[-1]
        
    plt.plot(index , good_points.values , 'y', label='original')
#         print('anomaly detected', original[-1])
#         plt.plot(original[-1] , 'r', label='original')

    plt.legend()
    fig.canvas.draw()
    fig.show()
    plt.pause(.1)
    index+=1
```
